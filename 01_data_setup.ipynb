```python
import pandas as pd
```

```python
#Importing the dataset that had the address mapping
address=pd.read_csv('add.csv',index_col=0)
```

```python
address.head()
```

```python
address_1=address.drop('destination_type', 1)
```

```python
#Importing the dataset that has the entire travel information
traffic=pd.read_csv('traffic.csv',index_col=0)
traffic.head()
```

```python
list(traffic)
```

```python
#traffic_1=traffic.drop('id', 1)
```

```python
#traffic_1.head()
```

```python
traffic_add = pd.merge(traffic, address_1, left_on='origin',right_on='id',how='left')
```

```python
traffic_add.head()
```

```python
traffic_add_1=traffic_add.rename(columns={"name": "origin_desc"})
```

```python
traffic_add_2 = pd.merge(traffic_add_1, address_1, left_on='destination',right_on='id',how='left')
```

```python
traffic_add_2.head()
```

```python
traffic_add_3=traffic_add_2.rename(columns={"name": "destination_desc"})
```

```python
traffic_add_3.head()
```

```python
traffic_add_4=traffic_add_3.drop(['origin','destination'],1)
```

```python
traffic_add_4.head()
```

```python
# new data frame with split value columns 
traffic_add_5 = traffic_add_4["departure_time"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
traffic_add_4["Date"]= traffic_add_5[0] 
  
# making separate last name column from new data frame 
traffic_add_4["Time"]= traffic_add_5[1] 
  
```
